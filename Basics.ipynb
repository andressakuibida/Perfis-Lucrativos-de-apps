# Há itens duplicados no dataset Google Play  Store

Há aplicativos que possuem mais de uma entrada no dataset. E para nossa análise, é importante ter só uma instância de cada.
Observe abaixo, onde constatamos 4 instancias do app Instagram e 2 do app Facebook.

```python
for app in android[1:]:
    name = app[0]
    if name == 'Instagram':
        print(app)
    if name == 'Facebook':
        print(app)
```

Observe que no caso do app Instagram e Facebook, visto acima, a diferença entre as instâncias reside no número de reviews.  O que sugere que os dados foram coletados em horas diferentes.
Para corrigir esse erro, vamos remover as duplicatas. Podemos usar o critério de que quem tiver maior número de reviews permanece.

Abaixo, contabilizamos e identificamos o número de duplicados no dataset.

```python
duplicados = []
unicos = []
    
for app in android[1:]:
    nome = app[0]
    
    if nome in unicos:
        duplicados.append(nome)
    else:
        unicos.append(nome)

print('Numero de apps duplicados:', len(duplicados), 
      "\n\nExemplos de apps duplicados:", duplicados[:15])  
```

# Removendo instâncias duplicadas

A seguir, iremos remover as entradas duplicadas, mantendo somente a entrada com maior número de review.

Vamos começar preenchendo um dicionário com o nome do app e seu respectivo número de review max. 

```python
reviews_max = {}

for app in android[1:]:

    nome = app[0]
    n_reviews = float(app[3])
    
    if nome in reviews_max and reviews_max[nome] < n_reviews:
        reviews_max[nome] = n_reviews
    elif nome not in reviews_max:
        reviews_max[nome] = n_reviews
        
print("Tamanho esperado: ", len(android[1:])- len(duplicados))
print("Tamanho obtido:", len(reviews_max))
```

Agora, usaremos o dicionário criado para realmente remover os duplicados de menor número de reviews.

Para isso, vamos percorrer os dados dos aplicativos do google e verificar se o número de reviews de cada app corresponde ao seu  máx numero de reviews e se essa instância já não se encontra na nova lista, se sim, a instância é adicionada a uma nova lista que corresponde aos dados já limpos, sem instâncias duplicadas.

```python
android_clean = []
adicionados = []

for app in android[1:]:
    nome = app[0]
    n_reviews = float(app[3])
    
    if (n_reviews == reviews_max[nome]) and (nome not in adicionados):
        android_clean.append(app)
        adicionados.append(nome)
        
print("Tamanho esperado:", len(reviews_max),
      "\nTamanho obtido:", len(android_clean),
      "\nTamanho dos dados não limpos:", len(android[1:]))        
```

# Removendo caracteres especiais

Nosso obejto de análise são apps direcionados á uma audiência de falantes de inglês. Se analisarmos o dataset, encontraremos apps que não são direcionados a falantes de inglês:

Textos em inglês incluem o alfabeto, números(0-9) e pontuação/símbolos (.,!,?,+,*,/) *.

Por trás dos panos, cada carácter que usamos em uma string possui um representante numeral, e a tabela que usamos para detalhar isso é a conhecida como tabela * **ASCII** (American Standard Code for Information Interchange) *.

Os números que correspondem aos carctéres comumente usados em um texto escrito em inglês variam de 0 á 127. Portanto, nós precisamos verificar se cada letra das strings no dataset está dentro desse range.

Para isso, nós temos a função * **ord( )** *, que nos retorna o número correspondente de cada caractere.

A seguir iremos criar uma função que verifica se uma determinada string é um texto inglês, ou seja, se está dentro do range 0-127:

```python
def is_english(string):
    for caracter in string:
        if ord(caracter) > 127:
            return False
    return True        
```

Agora iremos testar a função usando algumas strings:

```python
strings = ["Instagram",
           '爱奇艺PPS -《欢乐颂2》电视剧热播',
           'Docs To Go™ Free Office Suite',
           'Instachat 😜']

for string in strings:
    print(is_english(string))
```

Como pode ser visto acima, a função etiquetou apps em inglês como não-ingleses erroneamente, pois possuem emojis.

Para corrigir isso, só iremos etiquetar apps como não ingleses, se o app tiver mais do que 3 caractéres maiores que 127.

```python
def string_em_ingles (string):
   
    ''' 
    Retorna False se a string possuir
    mais que três caracteres especias. 
    Retorna True, caso contrário.
    '''
    
    i = 0
    for caracter in string:
        if ord(caracter) > 127:
            i += 1
            if i > 3:
                return False
    return True        
```

```python
for string in strings:
    
    print(string,": ", string_em_ingles(string),"\n")
```

A função criada para separar apps em inglês de não-inglês não é 100% eficiente, mas cumpre o papel da separação no nosso caso, como testao acima.

Abaixo, vamos retirar de ambos datasets, as instâncias que não são texto em inglês.

```python
#Retirando instancias com cacteres especiais

#ANDROID
android_cleaner = []

for app in android_clean:
    nome = app[0]
    if string_em_ingles(nome):
        android_cleaner.append(app)

#IOS
ios_clean = []

for app in ios:
    nome = app[1]
    if string_em_ingles(nome):
        ios_clean.append(app)
```

Agora, criarei uma função para auxiliar na construção de tabelas de frequencia, já que precisaremos construí-las para 3 colunas.

```python
def tabela_freq(dataset, indice):
    ''' 
    Retorna uma tabela de frequências dado uma lista de listas,
    e o índice da coluna que se quer a frequencia dos itens
    '''
    tabela = {}
    for row in dataset:
        tabela[row[indice]] = tabela.get(row[indice], 0) + 1
    return tabela


def display_table(dataset, indice):
    '''
    Mostra na tela a tabela de frequência da coluna índice, do 
    dataset *dataset*, de forma ordenada descendente
    '''
    table = tabela_freq(dataset, indice)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0],'->',
              round((entry[0]/len(dataset))*100, 3),"%" )
```

# Analisando o número de downloads por gênero

Aqui nós vamos calcular a média de downloads dos apps, considerando separadamente cada gênero. 

Vamos iniciar criando uma tabela de frequência para os gêneros

## Analisando IOS apps

```python
prime_genre = tabela_freq(ios_free,
              ios_header.index("prime_genre"))
dic = {}
for genre in prime_genre:
    total = 0
    len_genre = 0
    
    for app in ios_free:
        genre_app = app[ios_header.index("prime_genre")]
        
        if genre_app == genre:
            rating = float(app[ios_header.index('rating_count_tot')])
            total += rating
            len_genre += 1
            
    rating_avg = total/len_genre
    dic[genre] = rating_avg
    print("Gênero: ", genre, "| Média: ", rating_avg )
```

```python
table_display=[]
for key in dic:
    key_val_as_tuple = (dic[key], key)
    table_display.append(key_val_as_tuple)

table_sorted = sorted(table_display, reverse = True)
for entry in table_sorted:
    print(entry[1], ':', entry[0])
```

## Analisando apps ANDROID 

A coluna que possui o número de downloads dos apps do android possui caracteres especiais e para computar as médias, precisamos converte-los para *float*.

Após feito isso, vamos verificar quais categorias possuem maior numero de instalações.

```python
dic2 = {}
category = tabela_freq(android_free, 1)

for genre in category:
    total = 0
    len_category = 0
    
    for app in android_free:
        category_app = app[1]
        if genre == category_app:
            n_installs = app[5]
            n_installs = n_installs.replace("+","")
            n_installs = float(n_installs.replace(",",""))
            total += n_installs
            len_category += 1
            
    avg_installs = total/len_category
    dic2[genre] = avg_installs
    #print("Categoria: ", genre, "| Média: ", avg_installs )
```

```python
table_display=[]
for key in dic2:
    key_val_as_tuple = (dic2[key], key)
    table_display.append(key_val_as_tuple)

table_sorted = sorted(table_display, reverse = True)
for entry in table_sorted:
    print(entry[1], ':', entry[0])
```
