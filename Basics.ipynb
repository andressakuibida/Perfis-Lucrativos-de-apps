# H√° itens duplicados no dataset Google Play  Store

H√° aplicativos que possuem mais de uma entrada no dataset. E para nossa an√°lise, √© importante ter s√≥ uma inst√¢ncia de cada.

Observe que no caso do app Instagram e Facebook, visto acima, a diferen√ßa entre as inst√¢ncias reside no n√∫mero de reviews.  O que sugere que os dados foram coletados em horas diferentes.
Para corrigir esse erro, vamos remover as duplicatas. Podemos usar o crit√©rio de que quem tiver maior n√∫mero de reviews permanece.

Abaixo, contabilizamos e identificamos o n√∫mero de duplicados no dataset.

```python
duplicados = []
unicos = []
    
for app in android[1:]:
    nome = app[0]
    
    if nome in unicos:
        duplicados.append(nome)
    else:
        unicos.append(nome)

print('Numero de apps duplicados:', len(duplicados), 
      "\n\nExemplos de apps duplicados:", duplicados[:15])  
```

# Removendo inst√¢ncias duplicadas

A seguir, iremos remover as entradas duplicadas, mantendo somente a entrada com maior n√∫mero de review.

Vamos come√ßar preenchendo um dicion√°rio com o nome do app e seu respectivo n√∫mero de review max. 

```python
reviews_max = {}

for app in android[1:]:

    nome = app[0]
    n_reviews = float(app[3])
    
    if nome in reviews_max and reviews_max[nome] < n_reviews:
        reviews_max[nome] = n_reviews
    elif nome not in reviews_max:
        reviews_max[nome] = n_reviews
        
print("Tamanho esperado: ", len(android[1:])- len(duplicados))
print("Tamanho obtido:", len(reviews_max))
```

Agora, usaremos o dicion√°rio criado para realmente remover os duplicados de menor n√∫mero de reviews.

Para isso, vamos percorrer os dados dos aplicativos do google e verificar se o n√∫mero de reviews de cada app corresponde ao seu  m√°x numero de reviews e se essa inst√¢ncia j√° n√£o se encontra na nova lista, se sim, a inst√¢ncia √© adicionada a uma nova lista que corresponde aos dados j√° limpos, sem inst√¢ncias duplicadas.

```python
android_clean = []
adicionados = []

for app in android[1:]:
    nome = app[0]
    n_reviews = float(app[3])
    
    if (n_reviews == reviews_max[nome]) and (nome not in adicionados):
        android_clean.append(app)
        adicionados.append(nome)
        
print("Tamanho esperado:", len(reviews_max),
      "\nTamanho obtido:", len(android_clean),
      "\nTamanho dos dados n√£o limpos:", len(android[1:]))        
```

Textos em ingl√™s incluem o alfabeto, n√∫meros(0-9) e pontua√ß√£o/s√≠mbolos (.,!,?,+,*,/) *.

Por tr√°s dos panos, cada car√°cter que usamos em uma string possui um representante numeral, e a tabela que usamos para detalhar isso √© a conhecida como tabela * **ASCII** (American Standard Code for Information Interchange) *.

Os n√∫meros que correspondem aos carct√©res comumente usados em um texto escrito em ingl√™s variam de 0 √° 127. Portanto, n√≥s precisamos verificar se cada letra das strings no dataset est√° dentro desse range.

Para isso, n√≥s temos a fun√ß√£o * **ord( )** *, que nos retorna o n√∫mero correspondente de cada caractere.

A seguir iremos criar uma fun√ß√£o que verifica se uma determinada string √© um texto ingl√™s, ou seja, se est√° dentro do range 0-127:

```python
def is_english(string):
    for caracter in string:
        if ord(caracter) > 127:
            return False
    return True        
```

Agora iremos testar a fun√ß√£o usando algumas strings:

```python
strings = ["Instagram",
           'Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠',
           'Docs To Go‚Ñ¢ Free Office Suite',
           'Instachat üòú']

for string in strings:
    print(is_english(string))
```

Como pode ser visto acima, a fun√ß√£o etiquetou apps em ingl√™s como n√£o-ingleses erroneamente, pois possuem emojis.

Para corrigir isso, s√≥ iremos etiquetar apps como n√£o ingleses, se o app tiver mais do que 3 caract√©res maiores que 127.

```python
def string_em_ingles (string):
   
    ''' 
    Retorna False se a string possuir
    mais que tr√™s caracteres especias. 
    Retorna True, caso contr√°rio.
    '''
    
    i = 0
    for caracter in string:
        if ord(caracter) > 127:
            i += 1
            if i > 3:
                return False
    return True        
```

Agora, criarei uma fun√ß√£o para auxiliar na constru√ß√£o de tabelas de frequencia, j√° que precisaremos constru√≠-las para 3 colunas.

```python
def tabela_freq(dataset, indice):
    ''' 
    Retorna uma tabela de frequ√™ncias dado uma lista de listas,
    e o √≠ndice da coluna que se quer a frequencia dos itens
    '''
    tabela = {}
    for row in dataset:
        tabela[row[indice]] = tabela.get(row[indice], 0) + 1
    return tabela


def display_table(dataset, indice):
    '''
    Mostra na tela a tabela de frequ√™ncia da coluna √≠ndice, do 
    dataset *dataset*, de forma ordenada descendente
    '''
    table = tabela_freq(dataset, indice)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)

    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', entry[0],'->',
              round((entry[0]/len(dataset))*100, 3),"%" )
```

## Analisando apps ANDROID 

A coluna que possui o n√∫mero de downloads dos apps do android possui caracteres especiais e para computar as m√©dias, precisamos converte-los para *float*.

Ap√≥s feito isso, vamos verificar quais categorias possuem maior numero de instala√ß√µes.

```python
dic2 = {}
category = tabela_freq(android_free, 1)

for genre in category:
    total = 0
    len_category = 0
    
    for app in android_free:
        category_app = app[1]
        if genre == category_app:
            n_installs = app[5]
            n_installs = n_installs.replace("+","")
            n_installs = float(n_installs.replace(",",""))
            total += n_installs
            len_category += 1
            
    avg_installs = total/len_category
    dic2[genre] = avg_installs
    #print("Categoria: ", genre, "| M√©dia: ", avg_installs )
```

```python
table_display=[]
for key in dic2:
    key_val_as_tuple = (dic2[key], key)
    table_display.append(key_val_as_tuple)

table_sorted = sorted(table_display, reverse = True)
for entry in table_sorted:
    print(entry[1], ':', entry[0])
```
